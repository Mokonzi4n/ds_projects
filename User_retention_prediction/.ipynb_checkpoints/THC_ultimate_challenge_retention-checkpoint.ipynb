### Part 2 ‐ Experiment and metrics design

**Weekdays**

Ultimate Gotham - Active at night

Ultimate Metropolis - Active in the day

**Weekends**

Ultimate Gotham - Active at night and day

Ultimate Metropolis - Active at night and day

**1. Key Measure of Success of Experiment**

- Increase in the proportion of driver partners of Metropolis drivers in Gotham during the night or Gotham drivers in Metropolis during the day on weekdays.

This will be an indication that the toll bridge is being traversed from Gotham to Metropolis or vice versa.


**2. Description of Experiment Design**

 a.)  
 
   - Consider a random sample of driver partners in each city.
   
   - Ensure our sample size is reasonably large ≥ 30 for each city
   
   - Make sure the total population of the drivers in each city should be at least 300 to satisfy independence
 
   - Determine the proportion of drivers from Metropolis in Gotham during the night p1 (or from Gotham to Metropolis during the day p2) before the experiment is implemented.
   
   - implement toll reimbursement for a few months to get the average proportion of drivers from one city available in the other during their busy periods.
   
   
     
 b.)
 
   - Develop a hypothesis test composed of a null and alternate hypothesis
   
       null hypothesis : Reimbursement of tolls does not increase the proportion P of Metropolis drivers available in Gotham city at night or Gotham drivers in Metropolis during the day
       
       $$ Ho : P ≤ p1 $$
       $$ Ha : P > p1 $$
       
       OR
       
       $$ Ho : P ≤ P2 $$
       $$ Ha : P > P2 $$
       
   - Do a Z-test to determine the p-value. 
   
 c.)
 
  - Assume a significance level ⍺ = 0.05. If the p-value < ⍺  we reject the null hypothesis Ho and accept the alternate Ha. Else we fail to reject Ho
  
  - p-value < 0.05 implies we are confident that 95% of the time reimbursement of tolls will lead to an increase in the proportion of drivers available in each city.

### Fill Nans

Fill the Nan values with the means of the columns indicated. The `describe()` method above shows that 50th percentile of the columns mentioned below are $≥4.9$. Using their means of $4.78$ and $4.60$ would not change the dataset much.

```python
# Filling all the nan values of the avg_rating_by_driver and avg_rating_of_driver columns 
# with the means of the columns

cols = ['avg_rating_by_driver', 'avg_rating_of_driver']
ultimate_chal = ultimate_data.copy()
ultimate_chal[cols]=ultimate_chal[cols].fillna(round(ultimate_chal.mean(),1))
```

```python
# all the values are filled but for the missing phone types.

ultimate_chal.info()
```

### 6th Month Activity

  * To determine activity in the 6th month, we add 6 months to all sign up dates and then determine if the User's last trip date lies within the sixth month.

```python
# adding 6th month to the data set.

ultimate_chal['sixth_month'] = ultimate_chal.signup_date + pd.DateOffset(months=6)
ultimate_chal[['signup_date','sixth_month']].head()
```

```python
# Introducing a 6th month active column. If the last trip date falls in the 6th month, the user is active
# and 1 is entered into the column else 0 is entered

for idx, row in ultimate_chal.iterrows():
    if row.last_trip_date in pd.date_range((row.sixth_month - pd.DateOffset(months=1)), row.sixth_month):
            ultimate_chal.loc[idx,'sixth_month_active'] = 1
    else:
            ultimate_chal.loc[idx,'sixth_month_active'] = 0
    
```

### Predictive Analysis

- Exclude all datetime entries as these will not be relevant in our analysis
- Do one hot encoding of categorical data like cities and phones
- Use cross validation to determine to determine the accuracy of the different algorithms like Logistic regression, KNN, SVM, Decision Trees, Naive Bayes and Random Forests. Choose the algorithm with the highest accuracy to be used in the analysis

```python
ultimate_fld = ultimate_fld.select_dtypes(exclude=np.datetime64) # exlclude datetime columns
```

```python
ultimate_fld = pd.get_dummies(ultimate_fld) # One-hot-encoding using pandas
```

```python
ultimate_fld.columns
```

```python
# Prepare dataset to be used in models
X = ultimate_fld.drop(['sixth_month_active'], axis=1).values
x_columns = ultimate_fld.drop(['sixth_month_active'], axis=1).columns
y =  ultimate_fld.sixth_month_active.values

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)

models=[]
modl_name=[]
results=[]


models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC(gamma='auto')))
models.append(('RF', RandomForestClassifier(n_estimators=100)))

for name, model in models:
    kfold = KFold(n_splits=10, random_state=42)
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring='accuracy')
    results.append(cv_results)
    modl_name.append(name)
    print(f'{name} : {cv_results.mean()}, {cv_results.std()}')
```

### Algorithm Choice

From the tests above it shows that SVM is the most accurate of the algorithms. However, it was more resource intensive and took longer.

**Random Forest was chosen  since it was very close in accuracy and was less resource intensive. 
It is also capable of managing multicolinearity, is robust to outliers and as tree based model can handle variable ranges of features without requiring scaling** 

```python
RF=RandomForestClassifier(n_estimators=100)

kfold = KFold(n_splits=10, random_state=42)
cv_results = cross_val_score(RF, X_train, y_train, cv=kfold, scoring='accuracy')

RF.fit(X_train,y_train)

y_pred = RF.predict(X_test)

print('Mean of Cross validation results is',cv_results.mean())
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

### Plot ROC Curve

```python
from sklearn.metrics import roc_curve, roc_auc_score

# compute predicted probabilities
y_pred_prob = RF.predict_proba(X_test)[:,1] 

# Generate ROC curve values:

fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)


# Plot ROC curve
plt.rcParams["figure.figsize"] = (15,15)
plt.plot([0,1], [0,1], 'k--')

plt.plot(fpr,tpr)

plt.xlabel('False Positive Rate')

plt.ylabel('True Positive Rate')

plt.title('ROC Curve')

plt.show()



print(f'The Area Under Curve AUC:{roc_auc_score(y_test, y_pred_prob):.5f}')
```

The model is a good model with AUC about 78% of the curve.
